```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline  
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
Anime_df = pd.read_csv("Anime_df_Full_Floats.csv")
Anime_df.info()
```

```python
Anime_Status_df = Anime_df[["Anime_Name","Score","Status"]]
Anime_Status_df_Top_100 = Anime_Status_df.head(4000)
Anime_Status_df_Dum = pd.get_dummies(Anime_Status_df["Status"])
Anime_Status_df_Dum.sum()
```

```python
Anime_Status_df_Dum.sum().plot(kind="bar", color = ["green","blue","orange"], figsize = (10, 3),fontsize = 15, rot=0)
plt.xlabel('Status', fontsize = 15)
plt.ylabel('Number of Animes', fontsize = 15)
plt.show()
```

```python
Anime_Status_df_Dum.sum().plot(kind="pie",autopct = "%.2f%%", figsize = (7,7),fontsize = 15)
```

```python
Anime_df = pd.read_csv("Anime_df_Full_Floats.csv")
Anime_df["Genres"] = Anime_df["Genres"].str.replace(", ",",")
Anime_df.Source
```

```python
Anime_Source_df = Anime_df[["Anime_Name","Score","Source"]]
Anime_Source_df_Top_100 = Anime_Source_df.head(100)
Anime_Source_df_dum = pd.concat([Anime_Source_df["Source"].str.get_dummies(sep=",")])
Anime_Source_df_dum_And_Score = Anime_Source_df_dum.assign(Score = Anime_Source_df["Score"])
Anime_Source_df_Top_100_dum = pd.concat([Anime_Source_df_Top_100["Source"].str.get_dummies(sep=",")])
#Anime_Genres_df_Top_100_dum.sum().sort_values(ascending = False)
#Anime_Source_df_dum.sum().sort_values(ascending = False)
Anime_Source_df_dum.to_csv("Source Get_Dumm df.csv")
```

```python
Anime_Source_df_dum_And_Score.sum().sort_values(ascending = False)
```

```python
Anime_Source_df_dum.sum().sort_values().plot(kind="barh", figsize = (10, 15),fontsize = 15)
#plt.legend(numpoints=6,loc=6)
plt.xlabel('Number Of Animes',fontsize = 15)
plt.ylabel('Source',fontsize = 15)
plt.show()
```

```python
Anime_df = pd.read_csv("Anime_df_Full_Floats.csv")
Anime_df["Genres"] = Anime_df["Genres"].str.replace(", ",",")
Anime_df.info()
```

```python
Anime_Genres_df = Anime_df[["Anime_Name","Score","Genres"]]
Anime_Genres_df_Top_100 = Anime_Genres_df.head(100)
Anime_Genres_df_dum = pd.concat([Anime_Genres_df["Genres"].str.get_dummies(sep=",")])
Anime_Genres_df_dum_And_Score = Anime_Genres_df_dum
Anime_Genres_df_dum_And_Score = Anime_Genres_df_dum_And_Score.assign(score = Anime_Genres_df["Score"])
Anime_Genres_df_Top_100_dum = pd.concat([Anime_Genres_df_Top_100["Genres"].str.get_dummies(sep=",")])
#Anime_Genres_df_Top_100_dum.sum().sort_values(ascending = False)
```

```python
Anime_Genres_df_dum_And_Score.info()
```

```python
Anime_Genres_df_dum_And_Score.corr(method = "pearson").score.sort_values(ascending = False)
```

```python
Anime_Genres_df_dum.sum().sort_values().plot(kind="barh", figsize = (10, 15),fontsize = 15)
#plt.legend(numpoints=6,loc=6)
plt.xlabel('Number Of Animes',fontsize = 15)
plt.ylabel('Genre',fontsize = 15)
plt.show()
```
