```python
###########################################################
#####reading the CSV #####################################
##########################################################
Anime_df = pd.read_csv("Anime_df_Full.csv")
```

```python
Anime_df.Demographic.isnull().sum()
```

```python
###########################################################
#####deleting unnecessary columns##########################
###and too much Null Values###############################
##########################################################
Anime_df_Clean = Anime_df.drop(["Synonyms", "Japanese", "French","Type", "Theme", "Demographic"], axis=1)
```

```python
Anime_df_Clean.describe(include = "all")
```

```python
######################################
#####replacing missing data with NaN###
######################################
Anime_df_Clean = Anime_df_Clean.replace({"None found, add some": np.nan, "Unknown": np.nan, "?": np.nan, "Not available": np.nan,"/A":np.nan, "":np.nan})
```

```python
Anime_df_Clean.describe(include = "all")
```

```python
Anime_df_Clean.info()
```

```python
###########################################################
#deleting all the columns with less the 4k non-Null values#
###########################################################
Anime_df_Clean = Anime_df_Clean.drop(["Broadcast", "Producers", "Licensors"], axis=1)
```

```python
Anime_df_Clean.describe(include = "all")
```

```python
Anime_df_Clean.isna().sum()
```

```python
###########################################################
#fixing or deleting all the rows with Nan##################
###########################################################
Anime_df_Clean = Anime_df_Clean.rename(columns={'Score:':'Score',"Ranked:":"Ranked","Duration":"Duration_in_Min"})
Anime_df_Clean = Anime_df_Clean[Anime_df_Clean["Scored_By"].notna()]
Anime_df_Clean = Anime_df_Clean[Anime_df_Clean["Ranked"].notna()]
Anime_df_Clean = Anime_df_Clean[Anime_df_Clean["Duration_in_Min"].notna()]
#Anime_df_Clean_No_Nan = Anime_df_Clean.dropna()
#Anime_df_Clean_No_Nan.info()
```
